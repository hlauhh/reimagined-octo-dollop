```python
## Web Scrape Nike Discount items
# https://www.nike.com.hk/fa24midonm_md/list.htm?intpromo=PETP
```

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests as r
```

```python
html_nike = r.get(r'https://www.nike.com.hk/fa24midonm_md/list.htm?intpromo=PETP')

html_nike.content

html_nike_bs = bs(html_nike.content, 'html.parser')
html_nike_bs
```

```python
# html_nike_bs.find()
# html_nike_bs.find_all(class_='list-group')

# html_nike_bs.findAll(element, attribute key value pair)
nike_bs_product = html_nike_bs.find_all('dl', {"class" : "product_list_hover"})

nike_bs_product
```

```python
print(type(nike_bs_product))
print(nike_bs_product)
print(nike_bs_product[0])
```

```python
# Find the product info text
print(nike_bs_product[1].find('span', {"class" : "up"}).text) # product name
print(nike_bs_product[1].find('span', {"class" : "down"}).text) # product chinese name
print(nike_bs_product[1].find('dd',{"class" : "color666"}).text) # prices
print(nike_bs_product[1].find('span',{"class" : "origin_price"}).text) # origin price
print(nike_bs_product[1].find('span',{"class" : "origin_price"}).text.split('HK$')[-1]) # origin price
print(nike_bs_product[1].find('dd',{"class" : "color666"}).text.split('HK$')) # discount price
print(nike_bs_product[1].find('dd',{"class" : "color666"}).text.split('HK$')[-1]) # discount price
```

```python
row_arrs = []

for item in nike_bs_product:
    try:
        name = item.find('span', {"class" : "up"}).text # product name
        name_c = item.find('span', {"class" : "down"}).text # product chinese name
        ori_price = item.find('span',{"class" : "origin_price"}).text.split('HK$')[-1] # origin price
        dis_price = item.find('dd',{"class" : "color666"}).text.split('HK$')[-1] # discount price
    except:
        print(name = item.find('span', {"class" : "up"}).text)
    row_arrs.append([name,name_c,ori_price,dis_price])
    
# print(row_arrs)

df = pd.DataFrame(data=row_arrs, columns=['Name','Chinese Name','Origin_Price','Discount_Price'])
# print(df)

df['Discount_Price'] = df['Discount_Price'].str.replace(",","").astype(int)
df['Origin_Price'] = df['Origin_Price'].str.replace(",","").astype(int)
df['Discount'] = (df['Discount_Price']/ df['Origin_Price'] -1) *-1

df.to_csv('nike.csv', index=False)
df.to_excel('nike.xlsx', index=False)
```
